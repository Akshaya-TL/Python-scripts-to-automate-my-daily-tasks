```python
import pandas as pd
```

```python
qbank = pd.read_csv(r'qbankjava.csv')
```

```python
qbank.head()
```

```python
question = list(qbank['Question'])
options1 = list(qbank['Choice 1'])
options2 = list(qbank['Choice 2'])
options3 = list(qbank['Choice 3'])
options4 = list(qbank['Choice 4'])
answers = list(qbank['Correct Choice '])
reason = list(qbank['Explanation '])
```

```python
req = list(zip(question, options1, options2, options3, options4, answers, reason))
```

```python
!pip install python-docx
import docx
from docx import *
my_doc = docx.Document()
my_doc.add_paragraph("MCQs")
my_doc.save("C:/docsByMe/mcqs.docx")
```

```python
i = 1
for x in req[:181]:
    my_doc.add_paragraph("Q"+str(i)+str(x[0]))
    my_doc.add_paragraph("1. " + str(x[1]))
    my_doc.add_paragraph("2. " + str(x[2]))
    my_doc.add_paragraph("3. " + str(x[3]))
    my_doc.add_paragraph("4. " + str(x[4]))
    my_doc.add_paragraph("correct: " + str(int(x[5])))
    my_doc.add_paragraph("reason: " + str(x[6]))
    my_doc.save("C:/docsByMe/mcqs.docx")
    i += 1
```
